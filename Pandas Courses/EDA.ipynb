### The largest countries (Area)

```python
df.sort_values('Area (km2)', ascending=False)[['Country', 'Area (km2)']][:10]
```

### The largest country (Population 2022)   //Top10

```python
dff = df.sort_values('2022 Population', ascending=False)[['Country', '2022 Population', 'World Population Percentage']].head(10)
dff['World Population Percentage'] = dff['World Population Percentage'].apply(lambda x : f'{x} %')    # Add '%' Sign
dff
```

### Correlation

```python
df.corr(numeric_only=True)
```

### Heatmap Correlation

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
# Fix sizes
plt.rcParams['figure.figsize'] = (5, 5)
plt.show()
```

### Grouping

```python
df.groupby('Continent').mean(numeric_only=True).sort_values('World Population Percentage', ascending=False)
```

```python
df.columns
```

```python
continents = df.groupby('Continent').mean(numeric_only=True).sort_values('World Population Percentage', ascending=False)
continents = continents[df.columns[12:4:-1]]    # Reversing the order
continents.T.plot(kind='line')
plt.rcParams['figure.figsize'] = (15, 8)
```

```python
continents.boxplot(figsize=(15, 8))
```
