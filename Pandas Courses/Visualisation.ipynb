```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Data\Ice Cream Ratings.csv')
df
```

```python
df.set_index('Date', inplace=True)
df
```

## Choosing the style:

```python
plt.style.use('tableau-colorblind10')
plt.style.available
```

```python
df.plot()
```

```python
df.plot(
  title='Ice Cream Rating',
  kind='bar',
  stacked=True
)
```

```python
df.plot.barh(
  title='Ice Cream Rating',
  stacked=True,
)
```

```python
df['Flavor Rating'].plot(
  title='Flavor Rating',
  xlabel='Values',
  ylabel='Date',
  kind='bar',
  sharex=True,
)
```

```python
df.plot.scatter(
  x='Texture Rating',
  y='Flavor Rating',
  s=200,
  c='red'
)
```

```python
df.plot.hist(
  bins=20
)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(
  y='Flavor Rating',
  figsize=(5, 10)
)
```
