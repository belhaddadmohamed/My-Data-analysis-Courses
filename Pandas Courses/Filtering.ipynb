```python
import pandas as pd
```

```python
world_df = pd.read_csv('Data/world_population.csv')
world_df
```

#### Filtering;

```python
world_df[world_df['Rank'] < 6]
```

```python
# Same filter using .loc[]
world_df.loc[world_df['Rank'] < 6, : ]
```

```python
myCountries = ['Algeria', 'Bangladesh']
world_df[world_df['Country'].isin(myCountries)]
```

```python
world_df[world_df['Country'].str.contains(pat='[A-z]*ria')]
```

### loc[], iloc[], filter()

```python
world_df.loc[[1,6,26], 'Capital']
```

```python
world_df.iloc[[1,6,26], 3]
```

```python
world_df2 = world_df.set_index('Country')
world_df2
```

```python
world_df2.filter(items=["Algeria"], axis=0)
```

```python
world_df2.filter(like='eria', axis=0)
```

```python
world_df[world_df['Rank'] < 20].sort_values(by=['Continent'], ascending=False)
```

```python
world_df[world_df['Rank'] < 20].sort_values(by=['Continent', 'Country'], ascending=[False, True])
```
