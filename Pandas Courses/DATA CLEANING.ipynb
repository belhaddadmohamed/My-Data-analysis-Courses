```python
import pandas as pd
```

```python
df = pd.read_excel(r'Data\Customer Call List.xlsx')
df
```

## Data Cleaning

```python
# Drop Duplicates
# df[df.duplicated()]
df.drop_duplicates(inplace=True)
df
```

```python
# Drop the 'Not_Useful_Column' Column
df.drop(axis=1, columns='Not_Useful_Column', inplace=True)
```

```python
# df['Last_Name'] = df['Last_Name'].str.strip('123/._')
df['Last_Name'] = df['Last_Name'].replace(r'/|\.*|_', '', regex=True)
df
```

### Standardazation

```python
# Replace 'N/a' with Numpy Nan values
import numpy as np
df = df.replace('N/a', np.nan)
```

```python
# FORMAT THE PHONE_NUMBER : Replace anything not a letter or number by empty string / Split by -
df['Phone_Number'] = df['Phone_Number'].replace(r'[^A-z0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] +"-"+ x[3:6] +"-"+ x[6:10])
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
df
```

## Splitting Columns

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace(r'^Y$', 'Yes', regex=True)
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace(r'^N$', 'No', regex=True)
df
```

### FillNa()

```python
df.fillna('', inplace=True)
df
```

### Filtering by the People we can call

### Method1:

```python
contact_numbers_df = df[df['Do_Not_Contact'] == 'No']
contact_numbers_df = contact_numbers_df[contact_numbers_df['Phone_Number']!='']
contact_numbers_df
```

### Method2:

```python
for ind in df.index:    # .index: Iteration over rows / df: Iteration over columns
  if df.loc[ind, 'Do_Not_Contact'] == 'Yes' or df.loc[ind, 'Phone_Number'] == '':
    df.drop(ind, inplace=True)
df
```

```python
# Make empty as 'No'
for ind in df.index:
  if df.loc[ind, 'Do_Not_Contact'] == '':
    df.loc[ind, 'Do_Not_Contact'] = 'No'
df
```
